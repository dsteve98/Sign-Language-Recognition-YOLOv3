```
import argparse 
import sys
import pandas as pd

alphabet = list('ABCDEFGHIKLMNOPQSTUVWXYZ')   #BISINDO
# alphabet = list('ABCDEFGHIKLMNOPQRSTUVWXY')   #ASL

def letter_to_int(letter):
    return alphabet.index(letter)

def cmatrix(filename, thresh):

  confmat = [[0 for i in range(24)] for j in range(24)]
  za_wild = [0 for i in range(24)]
  za_truth = [0 for i in range(24)]
  za_detect = [0 for i in range(24)]
  pre_val = [0 for i in range(24)]
  rec_val = [0 for i in range(24)]
  detected = []
  
  f = open(filename)
  
  lines  = [line.rstrip("\n") for line in f.readlines()]
  total_tp = 0
  total_precision = 0
  total_recall = 0
  total_ground_truth = 0
  total_detection = 0
  found = True
  current_label = "A" 
  current_loc = "idk"
  lab1 = 0 #ground truth label
  lab2 = 0 #detected label
  detecting_any = True
  any_wild = False

  prev_line = ""
  for line in lines:
    line = line.replace('/', ' ')
    args = line.split(' ')
    if(args[0] == "Enter") :
      if(found == False):
        if(detecting_any == False):
          print("no detection: " + current_loc)
        else:
          print("not detecting truth class: " + current_loc)
      if(len(args) < 9):  #the number depends on len(args)
        break
      found = False
      current_label = args[12] #depends on location
      current_loc = args[10] + "/" + args[12] + "/" + args[13]
      lab1 = letter_to_int(current_label)
      total_ground_truth += 1
      za_truth[lab1] += 1
      detecting_any = False
    elif(args[0]!=''):
      if(args[0][1]==":" and int(args[1][:-1]) >= thresh):
        # print(int(args[1][:-1]))
        lab2 = letter_to_int(args[0][0])
        detecting_any = True
        total_detection += 1
        za_detect[lab2] += 1
        if(args[0][0]==current_label):
          # print("naisu, " + args[0][0])
          if(found == True):
            print(args[0][0] + " -dupe/wild, " + current_loc + args[1])
            za_wild[lab2] += 1
            any_wild = True
          else:
            confmat[lab1][lab2] += 1
            found = True
        else:
          print(args[0][0] + " -bad, " + current_loc + args[1])
          confmat[lab1][lab2] += 1
  for line in confmat:
    print(line)
  if(any_wild):
    print("---------------------------")
    print("wild detection")
    for i in range(24):
      if(za_wild[i] > 0):
        print(alphabet[i] + " Detected = " + str(za_wild[i]))

  print("---------------------------")
  print("Ground Truth : " + str(total_ground_truth) + "\t - Detected : " + str(total_detection))

  for i in range(24):
    pre_val[i] = confmat[i][i] / za_detect[i] #precision = TP / total detection of the class
    rec_val[i] = confmat[i][i] / za_truth[i]  #recall = TP / total truth
    total_tp += confmat[i][i]
    total_precision += pre_val[i]
    total_recall += rec_val[i]

  for i in range(24):
    print("class_id = " + str(i) + "\tname = " + alphabet[i] + "\t Detected = " + str(za_detect[i]) + ",TP = " + str(confmat[i][i]) + ",Truth = " + str(za_truth[i]) + ",Precision = " + format(pre_val[i], '.2f') + ",Recall = " + format(rec_val[i], '.2f'))

  print("Total True Positive = " + str(total_tp))
  avg_pre = total_precision/24
  print("Average Precision = " + str(avg_pre))
  avg_rec = total_recall/24
  print("Average Recall = " + str(avg_rec))
  accu = total_tp/total_detection
  print("Accuracy = " + str(accu))
  fone = 2 * (avg_pre * avg_rec) / (avg_pre + avg_rec)
  print("F1 Score = " + str(fone))

  
  for i in range(24):
    print(za_truth[i] - confmat[i][i])
```

Dengan Batasan Satu Deteksi

```
alphabet = list('ABCDEFGHIKLMNOPQSTUVWXYZ')

def letter_to_int(letter):
    return alphabet.index(letter)

def cmatrix(filename, thresh):
  confmat = [[0 for i in range(24)] for j in range(24)]
  za_truth = [0 for i in range(24)]
  za_detect = [0 for i in range(24)]
  pre_val = [0 for i in range(24)]
  rec_val = [0 for i in range(24)]
  za_fn = [0 for i in range(24)]
  detected = []
  total_tp = 0
  total_precision = 0
  total_recall = 0
  total_ground_truth = 0
  total_detection = 0
  total_fn = 0
  found = True
  current_label = "A" 
  current_loc = "idk"
  lab1 = 0 #ground truth label
  lab2 = 0 #detected label
  detecting_any = True
  f = open(filename)
  lines  = [line.rstrip("\n") for line in f.readlines()]
  
  for line in lines:
    line = line.replace('/', ' ')
    args = line.split(' ')
    if(args[0] == "Enter") :
      detected.sort(key=lambda x: x[1], reverse = True)
      if(len(detected) > 0):
        detecting_any = True
        lab2 = detected[0][0]
        total_detection += 1
        za_detect[lab2] += 1
        if(lab2==letter_to_int(current_label)):
          confmat[lab1][lab2] += 1
        else:
          print(lab2 + " -bad, " + current_loc + args[1])
          confmat[lab1][lab2] += 1
      if(detecting_any == False):
        print("no detection: " + current_loc)
      if(len(args) < 9):  #check last line
        break
      current_label = args[12]
      current_loc = args[10] + "/" + args[12] + "/" + args[13]
      lab1 = letter_to_int(current_label)
      total_ground_truth += 1
      za_truth[lab1] += 1
      detecting_any = False
      detected.clear()
    elif(args[0]!=''):
      if(args[0][1]==":" and int(args[1][:-1]) >= thresh):
        temp = letter_to_int(args[0][0]), int(args[1][:-1])
        detected.append(temp)

  for line in confmat:
    print(line)

  for i in range(24):
    pre_val[i] = confmat[i][i] / za_detect[i] #precision = TP / total detection of the class
    rec_val[i] = confmat[i][i] / za_truth[i]  #recall = TP / total truth
    za_fn[i] = za_truth[i] - confmat[i][i]
    total_tp += confmat[i][i]
    total_fn += za_fn[i]
    total_precision += pre_val[i]
    total_recall += rec_val[i]

  print("---------------------------")
  print("Ground Truth : " + str(total_ground_truth))
  print("Detected : " + str(total_detection))
  print("Total True Positive = " + str(total_tp))

  for i in range(24):
    print("class_id = " + str(i) + "\tname = " +
          alphabet[i] + "\t Detected = " +
          str(za_detect[i]) + ",TP = " +
          str(confmat[i][i]) + ",Truth = " +
          str(za_truth[i]) + ",Precision = " +
          format(pre_val[i], '.2f') + ",Recall = " +
          format(rec_val[i], '.2f'))

  avg_pre = total_precision/24
  print("Average Precision = " + str(avg_pre))
  avg_rec = total_recall/24
  print("Average Recall = " + str(avg_rec))
  accu = total_tp/(total_detection + total_fn)
  print("Accuracy = " + str(accu))
  fone = 2 * (avg_pre * avg_rec) / (avg_pre + avg_rec)
  print("F1 Score = " + str(fone))

  for i in range(24):
    print(za_truth[i] - confmat[i][i])
```
