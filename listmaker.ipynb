Train List

```
cd /gdrive/My\ Drive/darknet/bisindo
```

```
import glob, os
text_file = open("train_full2.txt", "w")
```

```
for clname in folders:
  print(clname)
  os.chdir("/gdrive/My Drive/darknet/dataset/bisindo/train/"+clname)
  for filename in glob.glob("*.jpg"):
    n = text_file.write("/content/gdrive/My Drive/darknet/dataset/bisindo/train/" + clname + "/" + filename + "\n")
  os.chdir("/gdrive/My Drive/darknet/dataset/bisindo_plus/train/"+clname)
  for filename in glob.glob("*.jpg"):
    n = text_file.write("/content/gdrive/My Drive/darknet/dataset/bisindo_plus/train/" + clname + "/" + filename + "\n")
```

```
text_file.close()
```
